```python
import os


image_dir = '/content/train/images'
label_dir = '/content/train/labels'


image_files = [f for f in os.listdir(image_dir) if f.endswith('.jpg') or f.endswith('.png')]
label_files = [f for f in os.listdir(label_dir) if f.endswith('.txt')]


print(f"Number of images: {len(image_files)}")
print(f"Number of label files: {len(label_files)}")


for image_file in image_files:
    label_file = os.path.splitext(image_file)[0] + '.txt'
    if label_file not in label_files:
        print(f"Missing label for image: {image_file}")


for label_file in label_files:
    image_file = os.path.splitext(label_file)[0] + '.jpg'  # or '.png'
    if image_file not in image_files:
        print(f"Missing image for label: {label_file}")
```

```python
# Check if each image has a corresponding label file
for image_file in image_files:
    label_file = os.path.splitext(image_file)[0] + '.txt'  # Get the corresponding label file name
    if label_file not in label_files:
        print(f"Missing label for image: {image_file}")
```
